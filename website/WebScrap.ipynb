```python
import requests
from bs4 import BeautifulSoup

# The URL of the webpage you want to scrape
url = 'https://marymount.edu/'

# Perform the request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the response using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Example: Extract the text from the first <h1> tag on the page
    h1_tag = soup.find('h1').text
    
    print(h1_tag)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the webpage you want to scrape
url = 'https://marymount.edu/'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Initialize BeautifulSoup to parse the content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Remove script and style elements
    for script_or_style in soup(["script", "style"]):
        script_or_style.decompose()

    # Get text and remove leading/trailing whitespace
    text = soup.get_text(separator=' ', strip=True)

    # Further clean text as needed
    cleaned_text = ' '.join(text.split())

    # Write the cleaned text to a file
    with open('cleaned_data.txt', 'w') as file:
        file.write(cleaned_text)

    print("Data written to cleaned_data.txt")
else:
    print(f"Error: Status code {response.status_code}")
```
